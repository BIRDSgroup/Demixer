```python
import csv
import numpy as np
import pandas as pd
import pickle
import matplotlib.pyplot as plt

with open('../scripts/ablation/invitro_cores_list1.pkl', 'rb') as file:
    data1 = pickle.load(file)
with open('../scripts/ablation/invitro_cores_list4.pkl', 'rb') as file:
    data4 = pickle.load(file)
with open('../scripts/ablation/invitro_cores_list5.pkl', 'rb') as file:
    data5 = pickle.load(file)
```

```python
cores=[1,2,4,8,16,32,64]
plt.figure(figsize=(6, 5))

P_values = ["no", "sampling_parallelization", ""]

plt.plot(cores, data1, label="no parallelization")
plt.plot(cores, data4, label="Sample level parallelization")
plt.plot(cores, data5, label="Sample level + SNP level parallelization")

plt.title("Time taken vs. Number of Cores (in vitro)",size=10)
plt.xlabel("Number of Cores",size=8)
plt.ylabel("Time taken (in seconds)",size=8)
plt.legend()
plt.grid(True)
plt.tick_params(axis='x', labelsize=8)
plt.tick_params(axis='y', labelsize=8)
plt.legend(fontsize=9)

plt.savefig('plot1.png', dpi=300)
plt.show()
```
